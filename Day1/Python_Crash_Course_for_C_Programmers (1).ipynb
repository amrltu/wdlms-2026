## 2. 📦 Variables & Data Types

**In C**, you must declare the type:
```c
int age = 20;
float pi = 3.14;
char grade = 'A';
char name[] = "Aatiz";
```

**In Python**, Python figures out the type automatically:

````python
age = 20          # int
pi = 3.14         # float
grade = 'A'       # str  (no separate char type in Python!)
name = "Aatiz"    # str
is_student = True # bool (True / False, capital T and F)

# Check what type a variable is
print(type(age))
print(type(pi))
print(type(name))
print(type(is_student))
````

````python
# Python allows multiple assignment in one line
x, y, z = 10, 20, 30
print(x, y, z)

# Swap two variables — no temp variable needed!
# In C: int temp = x; x = y; y = temp;
x, y = y, x
print("After swap:", x, y)
````

### Type Conversion
```c
// C casting
int x = (int) 3.9;  // → 3
```

````python
# Python type conversion
x = int(3.9)        # → 3
y = float(5)        # → 5.0
z = str(42)         # → "42"
a = int("100")      # → 100

print(x, y, z, a)
````

## 3. 🖥️ Input & Output

**In C:**
```c
int age;
printf("Enter your age: ");
scanf("%d", &age);
printf("You are %d years old.\n", age);
```

**In Python:**

````python
# input() always returns a STRING — convert if you need a number
age = int(input("Enter your age: "))
print("You are", age, "years old.")
````

````python
# f-strings: the modern Python way to format output (like printf)
name = input("Enter your name: ")
age = int(input("Enter your age: "))

# Method 1: f-string (recommended)
print(f"Hello {name}, you are {age} years old.")

# Method 2: .format()
print("Hello {}, you are {} years old.".format(name, age))

# Method 3: old-style % (like C's printf)
print("Hello %s, you are %d years old." % (name, age))
````

## 4. ➕ Operators

Most operators work the same as C. Here are the differences:

````python
# Arithmetic — same as C mostly
a, b = 17, 5

print(a + b)   # Addition      → 22
print(a - b)   # Subtraction   → 12
print(a * b)   # Multiplication→ 85
print(a / b)   # Division      → 3.4  (always float! unlike C int division)
print(a // b)  # Floor div     → 3    (like C int division)
print(a % b)   # Modulus       → 2
print(a ** 2)  # Exponent      → 289  (NO ^ in Python — ^ is XOR like in C)
````

````python
# Comparison operators — same as C
x = 10
print(x == 10)   # True
print(x != 5)    # True
print(x > 5)     # True
print(x <= 10)   # True
````

````python
# Logical operators — Python uses WORDS, not symbols!
# C:      &&    ||    !
# Python: and   or    not

x = 10
print(x > 5 and x < 20)   # True
print(x < 5 or x == 10)   # True
print(not (x == 10))       # False
````

## 5. 🔀 Conditional Statements

**In C:**
```c
if (marks >= 90) {
    printf("Grade A");
} else if (marks >= 75) {
    printf("Grade B");
} else {
    printf("Grade C");
}
```

**In Python — use `elif` instead of `else if`, and NO braces, use indentation:**

````python
marks = int(input("Enter your marks (0-100): "))

if marks >= 90:
    print("Grade A - Distinction!")
elif marks >= 75:
    print("Grade B - First Division")
elif marks >= 60:
    print("Grade C - Second Division")
elif marks >= 45:
    print("Grade D - Pass")
else:
    print("Fail")
````

````python
# Ternary / one-line if-else
# C:      int max = (a > b) ? a : b;
# Python:

a, b = 15, 22
max_val = a if a > b else b
print("Maximum:", max_val)
````

## 6. 🔄 Loops

### 6.1 — `while` loop (same logic as C)

````python
# C: int i = 1; while (i <= 5) { printf("%d ", i); i++; }

i = 1
while i <= 5:
    print(i, end=" ")   # end=" " prevents newline (like printf without \n)
    i += 1              # No i++ in Python! Use i += 1
````

### 6.2 — `for` loop (very different from C!)

Python's `for` loop is more like a **foreach** loop. Use `range()` to count.

````python
# C: for (int i = 0; i < 5; i++) { printf("%d ", i); }

for i in range(5):          # range(5) → 0, 1, 2, 3, 4
    print(i, end=" ")
print()  # newline

for i in range(1, 6):       # range(start, stop) → 1, 2, 3, 4, 5
    print(i, end=" ")
print()

for i in range(0, 20, 5):   # range(start, stop, step) → 0, 5, 10, 15
    print(i, end=" ")
````

````python
# Nested loops — Multiplication table
n = int(input("Enter a number for multiplication table: "))

for i in range(1, 11):
    print(f"{n} x {i:2} = {n*i:3}")
````

````python
# break and continue — same as C!

print("break example:")
for i in range(1, 10):
    if i == 5:
        break
    print(i, end=" ")

print("\n\ncontinue example:")
for i in range(1, 10):
    if i % 2 == 0:     # skip even numbers
        continue
    print(i, end=" ")
````

## 7. 🔧 Functions

**In C:**
```c
int add(int a, int b) {
    return a + b;
}
```

**In Python — use `def`, no type declarations:**

````python
# Basic function
def add(a, b):
    return a + b

result = add(10, 20)
print("Sum:", result)
````

````python
# Default parameters (not in C!)
def greet(name, message="Hello"):
    print(f"{message}, {name}!")

greet("Aatiz")               # uses default message
greet("Aatiz", "Namaste")    # custom message
````

````python
# Return multiple values (not possible in C without pointers/structs!)
def min_max(numbers):
    return min(numbers), max(numbers)

low, high = min_max([3, 1, 7, 2, 9, 4])
print(f"Min: {low}, Max: {high}")
````

````python
# Recursive function — same idea as C
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

for i in range(1, 8):
    print(f"{i}! = {factorial(i)}")
````

## 8. 📋 Lists (Python's Arrays)

**In C:**
```c
int arr[5] = {10, 20, 30, 40, 50};
```

Python **lists** are like arrays, but:
- Can hold **mixed types**
- **Dynamic size** — grow/shrink automatically
- **Powerful built-in methods**

````python
# Creating a list
marks = [85, 92, 78, 95, 88]

print(marks)           # whole list
print(marks[0])        # first element (index starts at 0, same as C)
print(marks[-1])       # LAST element (negative indexing — not in C!)
print(marks[1:4])      # slicing → [92, 78, 95] (index 1 to 3)
````

````python
# List operations
fruits = ["apple", "banana", "mango"]

fruits.append("orange")       # add to end
fruits.insert(1, "cherry")    # insert at position
fruits.remove("banana")       # remove by value
popped = fruits.pop()         # remove last element

print(fruits)
print("Popped:", popped)
print("Length:", len(fruits))  # len() — like sizeof(arr)/sizeof(arr[0]) in C
````

````python
# Iterating a list — Python's for loop shines here!
scores = [78, 92, 85, 67, 95, 74]

# Method 1: iterate values directly
for score in scores:
    print(score, end=" ")
print()

# Method 2: iterate with index (like C array loop)
for i in range(len(scores)):
    print(f"scores[{i}] = {scores[i]}")

# Method 3: enumerate (get both index and value)
for i, score in enumerate(scores):
    print(f"Student {i+1}: {score}")
````

````python
# Useful list functions
numbers = [3, 1, 7, 2, 9, 4, 6]

print("Sum:", sum(numbers))
print("Min:", min(numbers))
print("Max:", max(numbers))
print("Sorted:", sorted(numbers))   # returns new sorted list

numbers.sort()                       # sorts in-place
print("After sort:", numbers)
````

## 9. 🔤 Strings

In C, strings are character arrays (`char[]`). In Python, strings are **first-class objects** with many built-in methods — much easier!

````python
s = "Hello, Python!"

print(len(s))              # length → like strlen() in C
print(s.upper())           # → HELLO, PYTHON!
print(s.lower())           # → hello, python!
print(s.replace("Python", "World"))  # → Hello, World!
print(s[0:5])              # slicing → Hello
print(s.find("Python"))    # index where "Python" starts → 7
print("Python" in s)       # True — check if substring exists
````

````python
# Splitting and joining strings
sentence = "Nepal Army is professional"

words = sentence.split()        # splits by space → list of words
print(words)

joined = "-".join(words)        # join with separator
print(joined)
````

````python
# String iteration — like looping through char array in C
text = "Python"
for char in text:
    print(char, end="-")

# Count vowels example
print()
vowels = "aeiouAEIOU"
count = 0
for ch in text:
    if ch in vowels:
        count += 1
print(f"Vowels in '{text}':", count)
````

## 10. 📖 Dictionaries (Key-Value Store)

No direct C equivalent (closest is a struct or hash map). This is one of Python's most powerful features!

````python
# Dictionary = {key: value} pairs
student = {
    "name": "Aatiz",
    "age": 22,
    "marks": 85,
    "branch": "Computer Engineering"
}

# Access values
print(student["name"])           # → Aatiz
print(student.get("marks"))      # → 85  (safer — returns None if key missing)

# Add or update
student["city"] = "Kathmandu"    # add new key
student["marks"] = 90            # update existing

print(student)
````

````python
# Iterating a dictionary
for key, value in student.items():
    print(f"{key}: {value}")
````

## 11. 📁 File Handling

**In C:**
```c
FILE *fp = fopen("file.txt", "w");
fprintf(fp, "Hello!\n");
fclose(fp);
```

**Python uses `open()` and a `with` block (auto-closes file):**

````python
# WRITE to a file
with open("notes.txt", "w") as f:
    f.write("Python is easy!\n")
    f.write("No need to manage memory.\n")

print("File written successfully.")
````

````python
# READ from a file
with open("notes.txt", "r") as f:
    content = f.read()      # read entire file at once

print(content)
````

````python
# READ line by line
with open("notes.txt", "r") as f:
    for line in f:
        print(line.strip())   # strip() removes trailing \n
````

## 12. ⚠️ Error Handling

C uses error codes or `errno`. Python uses `try / except` (like try/catch in modern languages).

````python
# Division by zero crash prevention
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print(f"Result: {result:.2f}")

except ZeroDivisionError:
    print("❌ Error: Cannot divide by zero!")

except ValueError:
    print("❌ Error: Please enter valid numbers!")

finally:
    print("Program completed.")   # always runs, like cleanup in C
````

## 🏆 Mini Projects — Put It All Together!

### Project 1: Number Guessing Game

````python
import random

secret = random.randint(1, 100)
attempts = 0
max_attempts = 7

print("🎮 Guess the number (1-100)!")
print(f"You have {max_attempts} attempts.\n")

while attempts < max_attempts:
    guess = int(input(f"Attempt {attempts+1}/{max_attempts}: "))
    attempts += 1

    if guess == secret:
        print(f"🎉 Correct! You got it in {attempts} attempts!")
        break
    elif guess < secret:
        print("📈 Too low!")
    else:
        print("📉 Too high!")
else:
    print(f"😞 Game over! The number was {secret}.")
````

### Project 2: Student Grade Calculator

````python
def get_grade(avg):
    if avg >= 90:   return "A (Distinction)"
    elif avg >= 75: return "B (First Division)"
    elif avg >= 60: return "C (Second Division)"
    elif avg >= 45: return "D (Pass)"
    else:           return "F (Fail)"

subjects = ["Math", "Physics", "Chemistry", "English", "Computer"]
marks = []

print("📚 Student Grade Calculator")
print("-" * 30)

for subject in subjects:
    while True:
        try:
            m = float(input(f"Enter marks for {subject} (0-100): "))
            if 0 <= m <= 100:
                marks.append(m)
                break
            else:
                print("  ⚠ Marks must be between 0 and 100.")
        except ValueError:
            print("  ⚠ Please enter a valid number.")

average = sum(marks) / len(marks)
grade = get_grade(average)

print("\n" + "=" * 30)
print("       RESULT CARD")
print("=" * 30)
for i, sub in enumerate(subjects):
    print(f"{sub:<12}: {marks[i]:>6.1f}")
print("-" * 30)
print(f"{'Average':<12}: {average:>6.1f}")
print(f"{'Grade':<12}: {grade}")
print("=" * 30)
````

## 📝 Summary — C vs Python Quick Reference

| Concept | C | Python |
|---|---|---|
| Hello World | `printf("Hello");` | `print("Hello")` |
| Variable | `int x = 5;` | `x = 5` |
| Input | `scanf("%d", &x);` | `x = int(input())` |
| If-else | `if (x > 0) { }` | `if x > 0:` |
| else if | `else if` | `elif` |
| For loop | `for(i=0;i<n;i++)` | `for i in range(n):` |
| Array | `int arr[5]` | `arr = []` |
| String length | `strlen(s)` | `len(s)` |
| Logical AND | `&&` | `and` |
| Logical OR | `\|\|` | `or` |
| Logical NOT | `!` | `not` |
| Power | `pow(a,b)` | `a ** b` |
| Function | `int fn(int a) { }` | `def fn(a):` |
| File open | `fopen("f","r")` | `open("f","r")` |
| Error handling | `errno`, return codes | `try / except` |
| Increment | `i++` | `i += 1` |
| Null | `NULL` | `None` |

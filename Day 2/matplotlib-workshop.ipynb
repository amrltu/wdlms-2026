## 1. Importing & the Matplotlib Workflow

Matplotlib's standard import alias is `plt`. The recommended 5-step workflow for every plot:

| Step | Action |
|------|--------|
| 1 | **Prepare data** — arrays, model outputs |
| 2 | **Setup** — create `Figure` and `Axes` with `plt.subplots()` |
| 3 | **Plot** — call the right method on the `Axes` object (`ax.plot()`, `ax.scatter()`, …) |
| 4 | **Customize** — set title, axis labels, legend, limits |
| 5 | **Save / show** — `fig.savefig()` or `plt.show()` |

**Key vocabulary:**
- **`Figure`** (`fig`) — the entire canvas; may contain one or more Axes
- **`Axes`** (`ax`) — one individual plot on the Figure
- **`Axis`** — the x or y number line on an Axes

````python
# Full workflow in one block
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

fig, ax = plt.subplots(figsize=(6, 4))   # Step 2: setup
ax.plot(x, y)                             # Step 3: plot
ax.set(title="Sample Plot",               # Step 4: customize
       xlabel="x-axis",
       ylabel="y-axis");
# fig.savefig("plot.png")                 # Step 5: save (uncomment to use)
````

## 2. Two APIs: `pyplot` vs Object-Oriented

There are two main ways to build a plot:

| API | When to use | Example |
|-----|------------|--------|
| `plt.plot()` | Quick one-liners | `plt.plot(x, y)` |
| `fig, ax = plt.subplots()` | Full control, multiple axes | `ax.plot(x, y)` |

The **object-oriented API** (`plt.subplots()`) is the standard for ML work because it scales to multiple plots and is explicit about which Axes receives data.

````python
# pyplot style — simple but limited
plt.plot([1, 2, 3], [4, 5, 6]);
````

````python
# Object-oriented style — preferred
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])

print(type(fig))   # matplotlib.figure.Figure
print(type(ax))    # matplotlib.axes._axes.Axes
````

## 3. Core Plot Types with NumPy

Matplotlib visualizations are built on NumPy arrays. The four plot types you'll use most in ML:

| Plot | Method | ML use case |
|------|--------|-------------|
| Line | `ax.plot()` | Training/validation loss curves |
| Scatter | `ax.scatter()` | Feature relationships, cluster visualization |
| Bar | `ax.bar()` | Feature importance, class counts |
| Histogram | `ax.hist()` | Feature distributions, error distributions |

````python
# --- Line plot: trends over time / training curves ---
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.set(title="Line Plot", xlabel="x", ylabel="x²");
````

````python
# --- Scatter plot: individual data points / cluster visualisation ---
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
ax.set(title="Scatter Plot", xlabel="x", ylabel="sin(x)");
````

````python
# --- Bar plot: comparing categories / feature importance ---
features = ["age", "bmi", "glucose", "bp"]
importance = [0.35, 0.28, 0.22, 0.15]

fig, ax = plt.subplots()
ax.bar(features, importance)
ax.set(title="Feature Importance", ylabel="Importance score");
````

````python
# --- Histogram: distribution of a feature or error ---
np.random.seed(42)
errors = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(errors, bins=30)
ax.set(title="Model Error Distribution", xlabel="Error", ylabel="Count");
````

## 4. Multiple Subplots

`plt.subplots(nrows, ncols)` creates a grid of Axes — ideal for comparing multiple metrics side-by-side (e.g. different features or models at once).

`nrows × ncols = total number of Axes`

````python
# Option 1: unpack each Axes into its own named variable
np.random.seed(0)
x = np.linspace(0, 10, 100)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

ax1.plot(x, np.sin(x));                               ax1.set_title("Line")
ax2.scatter(np.random.rand(50), np.random.rand(50));  ax2.set_title("Scatter")
ax3.bar(["A", "B", "C"], [3, 7, 5]);                 ax3.set_title("Bar")
ax4.hist(np.random.randn(500));                       ax4.set_title("Histogram")

fig.suptitle("Four Plot Types", fontsize=14, fontweight="bold")
fig.tight_layout();
````

````python
# Option 2: index into the ax array — cleaner when many subplots
np.random.seed(1)
epochs = np.arange(1, 21)
train_loss = 1.0 / epochs + np.random.rand(20) * 0.05
val_loss   = 1.1 / epochs + np.random.rand(20) * 0.07
train_acc  = 1 - train_loss / 1.2
val_acc    = 1 - val_loss   / 1.2

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

ax[0].plot(epochs, train_loss, label="Train")
ax[0].plot(epochs, val_loss,   label="Val", linestyle="--")
ax[0].set(title="Loss", xlabel="Epoch", ylabel="Loss")
ax[0].legend()

ax[1].plot(epochs, train_acc, label="Train")
ax[1].plot(epochs, val_acc,   label="Val", linestyle="--")
ax[1].set(title="Accuracy", xlabel="Epoch", ylabel="Accuracy")
ax[1].legend()

fig.suptitle("Training Summary", fontsize=14, fontweight="bold")
fig.tight_layout();
````

## 5. Subplot Features: Shared Axes, Twin Axes & Annotations

| Feature | Parameter / Method | When to use |
|---------|-------------------|-------------|
| Shared x-axis | `sharex=True` | Multiple plots over the same time axis |
| Shared y-axis | `sharey=True` | Side-by-side histograms on the same scale |
| Twin y-axis | `ax.twinx()` | Two different scales on one plot (e.g. loss + accuracy) |
| Annotate a point | `ax.annotate()` | Highlight best epoch, outlier, threshold |

````python
# sharex=True: both subplots share the same x-axis — zooming one zooms both
np.random.seed(2)
epochs = np.arange(1, 31)
loss  = 2.0 * np.exp(-0.1 * epochs) + np.random.rand(30) * 0.05
acc   = 1 - loss / 2.2

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(8, 6))

ax0.plot(epochs, loss, color="tomato")
ax0.set(title="Loss", ylabel="Loss")

ax1.plot(epochs, acc, color="steelblue")
ax1.set(title="Accuracy", xlabel="Epoch", ylabel="Accuracy")

fig.suptitle("Training Metrics (shared x-axis)", fontsize=13, fontweight="bold")
fig.tight_layout();
````

````python
# sharey=True: compare distributions on the same y-scale
np.random.seed(3)
feature_a = np.random.randn(500) * 2 + 5
feature_b = np.random.randn(500) * 3 + 5

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(9, 4))

ax0.hist(feature_a, bins=30, color="steelblue")
ax0.set(title="Feature A", xlabel="Value", ylabel="Count")

ax1.hist(feature_b, bins=30, color="coral")
ax1.set(title="Feature B", xlabel="Value")

fig.suptitle("Feature Distributions (shared y-axis)", fontsize=13, fontweight="bold")
fig.tight_layout();
````

````python
# twinx(): plot two different metrics on one Axes with separate y-scales
fig, ax1 = plt.subplots(figsize=(8, 4))
ax2 = ax1.twinx()   # create a second Axes that shares the same x-axis

ax1.plot(epochs, loss, color="tomato",   label="Loss")
ax2.plot(epochs, acc,  color="steelblue", label="Accuracy", linestyle="--")

ax1.set(xlabel="Epoch", ylabel="Loss")
ax2.set_ylabel("Accuracy")

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="center right")

ax1.set_title("Loss & Accuracy on One Plot (twinx)");
````

````python
# ax.annotate(): highlight the best epoch
best_epoch = epochs[np.argmin(loss)]
best_loss  = loss.min()

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(epochs, loss, color="tomato")
ax.set(title="Training Loss", xlabel="Epoch", ylabel="Loss")

ax.annotate(
    f"Best: {best_loss:.3f}",
    xy=(best_epoch, best_loss),           # point to annotate
    xytext=(best_epoch + 3, best_loss + 0.15),  # label position
    arrowprops=dict(arrowstyle="->", color="black"),
    fontsize=10
);
````

## 6. Customizing Plots

Three main levers:

| What | How |
|------|-----|
| **Style** (overall theme) | `plt.style.use("ggplot")` — see `plt.style.available` |
| **Colors** | `cmap=` for continuous data; `color="r"` for single colors |
| **Labels & limits** | `ax.set(title=, xlabel=, ylabel=, xlim=, ylim=)` + `ax.legend()` |

````python
# See all available styles
print(plt.style.available)
````

````python
# Style: ggplot
x = np.linspace(0, 10, 100)

plt.style.use("ggplot")
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, np.sin(x), label="sin(x)")
ax.plot(x, np.cos(x), label="cos(x)", linestyle="--")
ax.set(title="ggplot style", xlabel="x", ylabel="y")
ax.legend();

plt.style.use("default")   # reset to default
````

````python
# Colormaps (cmap): colour scatter points by a continuous variable
np.random.seed(4)
n = 200
X = np.random.randn(n, 2)
y = X[:, 0]**2 + X[:, 1]**2    # distance from origin as a continuous label

fig, ax = plt.subplots(figsize=(6, 5))
scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap="plasma")
fig.colorbar(scatter, ax=ax, label="Distance²")
ax.set(title="Scatter with colormap", xlabel="x₁", ylabel="x₂");
````

````python
# Axis limits, legend position, mean line
np.random.seed(5)
ages    = np.random.randint(30, 80, 200)
chol    = 180 + ages * 0.8 + np.random.randn(200) * 20
disease = (chol > 240).astype(int)

fig, ax = plt.subplots(figsize=(8, 5))
scatter = ax.scatter(ages, chol, c=disease, cmap="RdYlGn_r")
ax.axhline(chol.mean(), color="navy", linestyle="--", label="Mean cholesterol")
ax.set(title="Cholesterol vs Age",
       xlabel="Age", ylabel="Cholesterol",
       xlim=[28, 82], ylim=[140, 320])
ax.legend(*scatter.legend_elements(), title="High chol.")
ax.legend(loc="upper left");
````
